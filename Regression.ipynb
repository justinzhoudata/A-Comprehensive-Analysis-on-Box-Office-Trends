```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import math
 

# Read the CSV file into a pandas dataframe
oscar_df = pd.read_csv('oscar.csv')
df = pd.read_csv('combineddf.csv')
```

```python
#Director Oscars

# Group the oscar dataframe by the name of the director and count the number of Oscars won
oscar_count_df = oscar_df.groupby('name').size().reset_index(name='oscars')

# Create a function that takes in a list of directors and returns the total number of Oscars won by them
def count_oscars(director_list):
    return oscar_count_df[oscar_count_df['name'].isin(director_list)]['oscars'].sum()

# Apply the count_oscars function to the "director" column in the "df" dataframe and store the result in a new column called "director_oscars"
df['director_oscars'] = df['director'].str.split(',').apply(count_oscars)

```

```python

# Create a function that takes in a list of crew members and returns the total number of Oscars won by them
def count_oscars_crew(crew_list):
    # Concatenate all the crew members into a single string
    crew_str = ','.join(crew_list)
    # Split the string into a list of individual crew members
    crew_members = crew_str.split(',')
    # Return the total number of Oscars won by the crew members
    return oscar_count_df[oscar_count_df['name'].isin(crew_members)]['oscars'].sum()

# Apply the count_oscars function to the "crew" column in the "df" dataframe and store the result in a new column called "crew_oscars"
df['crew_oscars'] = df['crew'].str.split(',').apply(count_oscars_crew)
```

```python
#Fix collection so that it is 1 or 0 if it is part of a collection or not. 
df['collection'] = df['collection'].fillna('').astype(bool).astype(int)
```

```python
#Save Coefficients
# Extract the coefficients for each independent variable
coefficients = lasso.coef_

# Create a list of the independent variable names
variable_names = list(X.columns)

# Create a dataframe with the variable names and coefficients
coefficients_df = pd.DataFrame({'Variable': variable_names, 'Coefficient': coefficients})

# Save the dataframe as a CSV file
coefficients_df.to_csv('lasso_coefficients.csv', index=False)
```
